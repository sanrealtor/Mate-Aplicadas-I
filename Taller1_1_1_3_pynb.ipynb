# ============. n.3. =============
Explica por qué...

$$ -\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n} $$

para \( n = 1, 2, 3, ...)

Sabemos que el valor de seno de (n) siempre está entre -1 y 1 para cualquier número real:

$$-1 \leq \sin(n) \leq 1 $$

Dividimos toda la desigualdad entre  n:

$$-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}$$

...

$$\lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{y} \quad \lim_{n \to \infty} \frac{1}{n} = 0$$
...

$$\lim_{n \to \infty} \frac{\sin(n)}{n} = 0$$

...

$$\lim_{n \to \infty} \frac{\sin(n)}{n} = 0$$
-------------------------------------------------


```python
# ============. n.4. =============

import numpy as np
import matplotlib.pyplot as plot

# (a)
a_values = [(1 + 1/n)**n for n in range(1, 6)]
for i, val in enumerate(a_values, start=1):
    print(f"a{i} = {val:.6f}")

# (b)
n_values = np.arange(1, 201)
an_values = (1 + 1/n_values) ** n_values

print("\nPrimeros 10 valores de aₙ:")
for i in range(10):
    print(f"a{i+1} = {an_values[i]:.6f}")

plot.figure(figsize=(10, 5))
plot.plot(n_values, an_values, color='orange', label='aₙ = (1 + 1/n)^n')
plot.axhline(y=np.e, color='red', linestyle='--', label='Número e ≈ 2.71828')
plot.xlabel('n')
plot.ylabel('aₙ')
plot.title('Comportamiento de aₙ = (1 + 1/n)^n')
plot.legend()
plot.grid(True)
plot.tight_layout()
plot.show()
```

# ============. n.13. =============

Se consideremos las siguientes sucesiones para encontrar:

$$a_n = n, \quad b_n = -n.$$

Límites:

$$\lim_{n \to \infty} a_n = \lim_{n \to \infty} n = \infty.$$

La sucesión $a_n$ **diverge a infinito**.

$$\lim_{n \to \infty} b_n = \lim_{n \to \infty} -n = -\infty.$$

La sucesión $b_n$ **diverge a menos infinito**.

Sumamos ambas sucesiones:

$$c_n = a_n + b_n = n + (-n) = 0.$$

Como la sucesión $c_n$ es siempre igual a 0, su límite es:

$$\lim_{n \to \infty} c_n = 0.$$

La sucesión $c_n$ es constante y converge a 0

Hemos encontrado dos sucesiones divergentes cuya suma es convergente.
